```python
import os
import pandas as pd
import yfinance as yf

# Function to download stock data and handle missing tickers
def fetch_stock_data(tickers, exchange_suffix, exchange_name):
    failed_tickers = []
    
    # Create a folder for the exchange
    os.makedirs(exchange_name, exist_ok=True)
    
    for ticker in tickers:
        full_ticker = ticker + exchange_suffix
        try:
            # Fetch historical data from the start date
            stock_data = yf.download(full_ticker, period="max")
            
            # If no data is returned, record the ticker in the failed list
            if stock_data.empty:
                failed_tickers.append(ticker)
            else:
                # Save the data to CSV in the respective exchange folder
                file_path = os.path.join(exchange_name, f"{ticker}.csv")
                stock_data.to_csv(file_path)
                print(f"Data for {ticker} saved successfully.")
        except Exception as e:
            print(f"Failed to fetch data for {ticker}: {str(e)}")
            failed_tickers.append(ticker)
    
    # Save the failed tickers into a separate file
    if failed_tickers:
        failed_tickers_df = pd.DataFrame(failed_tickers, columns=["Failed_Tickers"])
        failed_tickers_df.to_csv(f"failed_{exchange_name}_tickers.csv", index=False)
        print(f"Failed tickers saved to failed_{exchange_name}_tickers.csv")

# Load tickers from NSE and BSE files
nse_file_path = 'NSE_tickers.csv'  # Replace with your actual NSE file path
bse_file_path = 'bse_tickers.csv'  # Replace with your actual BSE file path

nse_df = pd.read_csv(nse_file_path)
bse_df = pd.read_csv(bse_file_path)

# Assuming the tickers are in a column named 'SYMBOL' in both files
nse_tickers = nse_df['SYMBOL'].tolist()
bse_tickers = bse_df['SYMBOL'].tolist()

# Fetch and save historical data for NSE and BSE tickers
fetch_stock_data(nse_tickers, '.NS', 'NSE')
fetch_stock_data(bse_tickers, '.BO', 'BSE')
```
